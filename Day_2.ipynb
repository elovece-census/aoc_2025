```python
with open('day_2_puzzle_input.txt') as f:
    all_ids = [id.split(',') for id in f]
    
all_ids = all_ids[0]
```

```python
all_ids
```

PART 1

```python
ids = []
for id in all_ids:
    start, end = map(int, id.split('-'))
    ids.append((start, end))
```

```python
ids
```

```python
all_invalid_ids = []
for start, end in ids:
    invalid_ids = []
    for num in range(start, end + 1):
        s = str(num)
        length = len(s)
        
        # Must have even length to be repeated twice
        if length % 2 != 0:
            continue
        
        # Check if first half equals second half
        mid = length // 2
        if s[:mid] == s[mid:]:
            invalid_ids.append(num)
    
    all_invalid_ids.extend(invalid_ids)
    if invalid_ids:
        print(f"{start}-{end} has invalid IDs: {invalid_ids}")
```

```python
# Sum them up
sum(all_invalid_ids)
```

PART 2

```python
all_invalid_ids = []
for start, end in ids:
    invalid_ids = []
    
    for num in range(start, end + 1):
        s = str(num)
        length = len(s)
    
        # Try all possible pattern lengths (from 1 to length//2)
        # Pattern must repeat at least twice, so max pattern length is length//2
        for pattern_len in range(1, length // 2 + 1):
            # Check if this pattern length divides evenly into total length
            if length % pattern_len == 0:
                pattern = s[:pattern_len]
                num_repetitions = length // pattern_len
                
                # Check if repeating the pattern gives us the original string
                if pattern * num_repetitions == s:
                    invalid_ids.append(num)
                    break
                
    all_invalid_ids.extend(invalid_ids)
    if invalid_ids:
        print(f"{start}-{end} has invalid IDs: {invalid_ids}")
```

```python
# Sum them up!!
sum(all_invalid_ids)
```
