```python
with open('day_4_puzzle_input.txt') as f:
    grid = [list(line.strip()) for line in f]
```

```python
grid
```

```python
viz_grid = [row[:] for row in grid]
viz_grid
```

PART 1

```python
# Parse the grid
rows = len(grid)
cols = len(grid[0])
```

```python
# Define 8 adjacent directions (including diagonals)
directions = [
    (-1, -1), (-1, 0), (-1, 1),  # top-left, top, top-right
    (0, -1), (0, 1),             # left, right
    (1, -1), (1, 0), (1, 1)      # bottom-left, bottom, bottom-right
]
```

```python
accessible_count = 0
accessible_positions = []
```

```python
# Check each position in the grid
for r in range(rows):
    for c in range(cols):
        if grid[r][c] == '@':
            # Count adjacent rows
            adjacent_rolls = 0
            for dr, dc in directions:
                nr, nc = r + dr, c + dc
                # Check if neighbor is within bounds and is a roll
                if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == '@':
                    adjacent_rolls += 1
                    
            # A roll is accessible if it has fewer than 4 adjacent rolls
            if adjacent_rolls < 4:
                accessible_count += 1
                accessible_positions.append((r, c))
                viz_grid[r][c] = 'x'
```

```python
accessible_count
```

```python
viz_grid = list('\n'.join(''.join(row) for row in viz_grid).split('\n'))
viz_grid
```

PART 2

```python
# Count total rolls that can be removed through iterative process.
# After removing accessible rolls, check for newly accessible rolls.

total_removed = 0
iteration = 0
```

```python
while True:
    # Find all accessible rolls in current state
    accessible = []
    
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == '@':
                # Count adjacent rolls
                adjacent_rolls = 0
                for dr, dc in directions:
                    nr, nc = r + dr, c + dc
                    # Check if neighbor is within bounds and is a roll
                    if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == '@':
                        adjacent_rolls += 1
                        
                # A roll is accessible if it has fewer than 4 adjacent rolls
                if adjacent_rolls < 4:
                    accessible.append((r, c))
                    
    # If no rolls accessible, we are done
    if not accessible:
        final_grid = grid
        break
    
    # Remove all accessible rolls
    for r, c in accessible:
        grid[r][c] = '.'
        
    iteration += 1
    total_removed += len(accessible)
    print(f"Iteration {iteration}: Removed {len(accessible)} rolls (total: {total_removed})")
```
