```python
with open('day_6_puzzle_input.txt') as f:
    lines = [line.strip().split('\n') for line in f]
    for i, _ in enumerate(lines):
        lines[i] = lines[i][0].split()
```

## PART 1

```python
# Identify problems by taking each set of numbers col-by-col and finding the associated operation
problem_cols = []
```

```python
for col_idx, _ in enumerate(lines[0]):
    column = [line[col_idx] for line in lines]
    problem_cols.append(column)
```

```python
problem_cols[:20]
```

```python
results = []
```

```python
# Calculate result based on operation
for problem in problem_cols:
    operation = problem[-1]
    numbers = [int(num) for num in problem[:-1]]
    
    if operation == '*':
        result = 1
        for num in numbers:
            result *= num
    else:  # operation == '+'
        result = sum(numbers)
    
    results.append(result)
```

```python
sum(results)
```
