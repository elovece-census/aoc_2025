```python
with open('day_1_puzzle_input.txt') as f:
    rotations = [line.strip() for line in f]
```

PART 1

```python
position = 50  # Starting position
zero_count = 0  # Number of times we land on zero
```

```python
for rotation in rotations:
    direction = rotation[0]  # L or R
    distance = int(rotation[1:])  # Number of clicks
    
    if direction == 'R':
        # Right rotation: add mod 100
        position = (position + distance) % 100
    else:  # direction == 'L'
        # Left rotation: subtract mod 100
        position = (position - distance) % 100
        
    # Check if we landed on zero
    if position == 0:
        zero_count += 1
```

PART 2

```python
position = 50  # Starting position
zero_count = 0  # Number of times we land on zero
```

```python
for rotation in rotations:
    direction = rotation[0]  # L or R
    distance = int(rotation[1:])  # Number of clicks
    
    # Count zeros in this rotation
    # We visit positions: position+1, position+2, ..., position+distance (for R)
    # Or: position-1, position-2, ..., position-distance (for L)
    
    if direction == 'R':
        # Moving right: positions are (position + 1) % 100, (position + 2) % 100, ..., (position + distance) % 100
        # We hit 0 when (position + k) % 100 == 0, i.e., when position + k ≡ 0 (mod 100)
        # This means k ≡ -position (mod 100)
        # We need k such that: 1 <= k <= distance and (position + k) % 100 == 0
        
        # First zero is at k = (100 - position) % 100, but k must be >= 1
        first_zero = (100 - position) % 100
        if first_zero == 0:
            first_zero = 100  # Retains mod 100 value while satisfying rule
            
    else:  # direction == 'L'
        # Moving left: positions are (position - 1) % 100, (position - 2) % 100, ..., (position - distance) % 100
        # We hit 0 when (position - k) % 100 == 0, i.e., when position - k ≡ 0 (mod 100)
        # This means k ≡ position (mod 100)
            
        first_zero = position % 100
        if first_zero == 0:
            first_zero = 100  # Retains mod 100 value while satisfying rule

    if distance >= first_zero:
        # We reach at least one zero
        # Count: 1 + how many more times we loop through
        zero_count += 1 + (distance - first_zero) // 100
            
    # Update position
    if direction == 'R':
        position = (position + distance) % 100
    else:
        position = (position - distance) % 100
```
