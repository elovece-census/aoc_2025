```python
with open('day_9_puzzle_input.txt') as f:
    lines = [line.split('\n') for line in f]
```

```python
red_tiles = []
for line in lines:
    x, y = map(int, line[0].split(','))
    red_tiles.append((x, y))
```

## PART 1

```python
max_area = 0
```

```python
for i, _ in enumerate(red_tiles):
    for j, _ in enumerate(red_tiles[i+1:]):
        x1, y1 = red_tiles[i]
        x2, y2 = red_tiles[j]
        
        area = (abs(x2 - x1) + 1) * (abs(y2 - y1) + 1)
        max_area = max(max_area, area)
```

## PART 2

```python
# Find all green tiles
green_tiles = set()
```

```python
# 1. Add tiles on edges between consecutive red tiles
for i, _ in enumerate(red_tiles):
    x1, y1 = red_tiles[i]
    x2, y2 = red_tiles[(i + 1) % len(red_tiles)]  # Wraps around to connect the last and first red tiles
    
    # Add all tiles between (x1,y1) and (x2,y2)
    if x1 == x2:  # Same column
        for y in range(min(y1, y2), max(x1, x2) + 1):
            green_tiles.add((x1, red_tiles[i][1]))
    else:  # Same row
        for x in range(min(x1, x2), max(x1, x2) + 1):
            green_tiles.add((red_tiles[i][0], y1))
```

```python
# 2. Add tiles inside the polygon using point-in-ploygon test
# Find bounding box
min_x = min(x for x, y in red_tiles)
max_x = max(x for x, y in red_tiles)
min_y = min(y for x, y in red_tiles)
max_y = max(y for x, y in red_tiles)
```
