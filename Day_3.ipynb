```python
with open('day_3_puzzle_input.txt') as f:
    numbers = [num.strip() for num in f]
```

```python
numbers
```

PART 1

```python
total_joltage = 0
```

```python
# Find max by checking largest digits first

for num in numbers:
    max_joltage = 0
    # For each position, find the best second digit after it
    for i, _ in enumerate(num):
        first_digit = num[i]
        # Find the maximum digit after position i
        if i + 1 < len(num):
            max_second = max(num[i+1:])
            joltage = int(first_digit + max_second)
            max_joltage = max(max_joltage, joltage)
            
    print(f"{num} -> {max_joltage}")
    total_joltage += max_joltage
```

```python
total_joltage
```

PART 2

```python
from collections import deque
```

```python
total_joltage = 0
```

```python
# Find the lexicographically largest subsequence of length 12
# while maintaining the relative order of digits

for num in numbers:
    to_remove = len(num) - 12
    
    stack = deque()
    
    for i, digit in enumerate(num):
        # Remove smaller digits from the stack if:
        # 1. We still need to remove digits (to_remove > 0)
        # 2. the current digit is larger than the top of the stack
        while stack and to_remove > 0 and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
            
        stack.append(digit)
        
    while to_remove > 0:
        stack.pop()
        to_remove -= 1
        
    max_joltage_str = ''.join(stack)
    max_joltage = int(max_joltage_str)
    print(f"{num} -> {max_joltage_str}")
    total_joltage += max_joltage
```
